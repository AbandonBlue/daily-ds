## 有效率地探索資料
**情境**
- 在小資料集時，可以透過人工檢查資料集，對資料集有其理解，但在資料巨大的時候，就是一個不可能的方法了。

**解法**
- 透過Clustering方法，將資料集進行分組，因為分群會有**組內差異小、組間差異大**的性質，相當於統計的代表性樣本的概念。

**注意**
- 集群數量?
- 集群間差異大嗎?
- 是否有些集群特別稀疏或者密集，稀疏集群，可能需要增加特徵或者資料幫助資料更容易處理。
- 有特別的pattern嗎? 需要回頭檢驗。

```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer(as_frame=True)
X = data['data']
y = data['target']
X, y
```

```python
X.describe()
```

> 可以發現各特徵的尺度不一，對於距離相關的模型會造成重大影響。

```python
# 在分類情況中，很自然會把集群數設定為分類數
# 若分群法依賴於距離，記得需要將特徵標準化
from sklearn.cluster import KMeans

k = 2
kmeans = KMeans(n_clusters=k, random_state=222)
```

```python
# 標準化特徵

from sklearn.preprocessing import StandardScaler

std = StandardScaler()
X_std = std.fit_transform(X)
X.iloc[:, :] = X_std
X.describe()
```

```python
kmeans.fit(X)
cluster_y = kmeans.predict(X)


cluster_y
```

```python
# 集群1的特徵

X[cluster_y==1].describe()
```

```python
# 集群的特徵

X[cluster_y==0].describe()
```

> 這邊只關注特徵的均值，可以看到是有明顯的差異，要透過數據佐證，可以透過t、z檢定去檢測，這邊先忽略

```python
# 查看標籤集群是否大致相符

is_cluster_zero = (cluster_y == 0)
is_label_one = (y == 1)
```

```python
(is_cluster_zero & is_label_one).sum()
```

```python
((~is_cluster_zero) & (~is_label_one)).sum()
```

```python
# 標籤不同者

num_label_not_eq = (len(X) - 339 - 176)
print(num_label_not_eq)
print(num_label_not_eq / len(X) * 100, '%')
```

> 大約是9%的標籤不同，初步判斷分群方法與標籤的pattern初步相同

> 有時候甚至可以將分群當作一個新的特徵標籤，幫助模型效果

## 搭配降維圖示理解
- 將特徵降為2d，畫於圖上直觀理解


```python
import umap.umap_ as umap

embeddings = X.values

reducer = umap.UMAP()
embedding = reducer.fit_transform(embeddings)
embedding.shape
```

```python
# 在真實標籤的觀察下

plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=["red" if e == 1 else "blue" for e in y ], 
)
```

```python
# 在分群標籤的觀察下

plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=["red" if e == 0 else "blue" for e in cluster_y ], 
)
```

> 可以發現觀察到的結果是很雷同的，某方面也顯現分群效果不錯，中間比較容易交錯的點也是算法容易出錯的地方。
