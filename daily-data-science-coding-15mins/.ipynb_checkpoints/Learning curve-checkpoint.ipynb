## Learning Curve
- 常用以觀察`訓練樣本數`與`模型效能(validation)`的圖形，可以得知 `Overfitting 的現象`以及`最佳使用樣本數`。

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## 1. 取得資料

data = load_breast_cancer()
X = data['data']
y = data['target']
```

```python
## 2. 切分資料

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
## 3. 訓練隨著樣本數增加

train_metrics = []
validation_metrics = []
sample_sizes = []

for i in range(10, len(x_train)):
    sample_sizes.append(i)
    x_temp = x_train[:i]
    y_temp = y_train[:i]
    
    rf = RandomForestClassifier()
    rf.fit(x_temp, y_temp)
    
    y_train_pred = rf.predict(x_train)
    acc_train = accuracy_score(y_train, y_train_pred)
    y_test_pred = rf.predict(x_test)
    acc_test = accuracy_score(y_test, y_test_pred)
    
    train_metrics.append(acc_train)
    validation_metrics.append(acc_test)
    
```

```python
## 4. 畫出 learning curve

plt.figure(figsize=(18, 10))
plt.plot(sample_sizes, train_metrics, color='blue')
plt.plot(sample_sizes, validation_metrics, color='red')
```

> 可以看到用一個簡單的癌症檢驗資料搭配隨機森林模型看到，大約樣本數超過`200`之後，就有些 overfitting 了 !
