```python
# import the TrendReq method from the pytrends request module
from pytrends.request import TrendReq
from pprint import pprint
# execute the TrendReq method by passing the host language (hl) and timezone (tz) parameters



pytrend = TrendReq(hl='en-US', tz=360)
keywords = ['Python', 'Java', 'C++']
pytrend.build_payload(
     kw_list=keywords,         # 搜尋字
     cat=0,                    # 類別
     timeframe='today 1-m',    # 時間區段
     geo='TW',                 # 地區
     gprop='')                 # What Google property to filter to


# 有多種, 可以去看文檔找自己需要的用法。
pprint(pytrend.interest_over_time())   # Request data from Google's Interest Over Time section and return a dataframe
```

> isPartial: 代表當天還沒過去

```python
# 試著爬取2020-現在的資料科學/機器學習/深度學習資料

# 建立物件
pytrend = TrendReq()

# 搜尋關鍵字
keywords = ['資料科學', '機器學習', '深度學習', '推薦系統', 'NLP']

# payload
pytrend.build_payload(
    kw_list=keywords,
    cat=0,
    timeframe='2020-01-01 2021-05-26',
    geo='TW',
    gprop=''
)

# 隨著時間搜尋
df = pytrend.interest_over_time()
```

```python
def crawler_pytrend(**kwargs):
    pytrend = TrendReq()
    keywords = kwargs['keywords']
    
    pytrend.build_payload(
        kw_list=keywords,
        cat=kwargs['cat'],
        timeframe=kwargs['timeframe'],
        geo=kwargs['geo'],
        gprop=kwargs['gprop']
    )
    # 所有method, 應該可以用State pattern來寫, 待完成...
    if kwargs['method'] == 'interest_over_time':
        df = pytrend.interest_over_time()
    
    return df
    
    

    
pytrend_config = {
    'keywords': ['資料科學', '機器學習', '深度學習', '推薦系統', 'NLP'],
    'cat': 0,
    'timeframe': 'today 12-m', # 從今天往前數12個月
    'geo': 'TW',
    'gprop': '',
    'method': 'interest_over_time'
}

df = crawler_pytrend(**pytrend_config)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


for c_name in df.columns[:-1]:
    plt.plot(df[c_name], label=c_name)  # 設定

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 中文
plt.legend(loc='best')  # 顯示legend
```
